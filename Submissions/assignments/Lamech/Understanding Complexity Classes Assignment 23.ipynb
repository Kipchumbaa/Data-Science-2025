```python
# Common imports and helpers
import timeit
import random
import math
import matplotlib.pyplot as plt
from datetime import datetime

# helper to run timeit for a callable (no-arg) and return seconds
def time_callable(func, number=10):
    t = timeit.timeit(func, number=number)
    return t/number

# helper to plot results
def plot_results(xs, ys, label=None, xlabel='n', ylabel='time (s)', title=None, logscale=False):
    plt.figure(figsize=(7,4))
    if logscale:
        plt.loglog(xs, ys, marker='o', label=label)
    else:
        plt.plot(xs, ys, marker='o', label=label)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if title:
        plt.title(title)
    if label:
        plt.legend()
    plt.grid(True)
    plt.show()
```

## Θ(1) — Constant Time Tasks

For each task: implement function, predict complexity, test on different input sizes (where relevant), and explain the results.

```python
# Task Θ(1) - Task 1: Access the first element in a list of 1M numbers

def access_first(L):
    return L[0]

# Prediction: Θ(1) — direct index access

# Prepare input (1 million elements)
L1 = list(range(10**6))

# Time the access (wrap in lambda to be no-arg for timeit)
print('Access first element:')
print('Result sample:', access_first(L1))
print('Time (avg over 1000 runs):', time_callable(lambda: access_first(L1), number=1000))
```

```python
# Task Θ(1) - Task 2: Check if a number is even using modulus

def is_even(x):
    return x % 2 == 0

# Prediction: Θ(1) — single arithmetic operation

print('\nIs even check:')
print('Sample:', is_even(123456))
print('Time (avg over 100000 runs):', time_callable(lambda: is_even(123456), number=100000))
```

```python
# Task Θ(1) - Task 3: Swap two variables

def swap(a, b):
    a, b = b, a
    return a, b

# Prediction: Θ(1)
print('\nSwap variables:')
print('Sample:', swap(1,2))
print('Time (avg over 100000 runs):', time_callable(lambda: swap(1,2), number=100000))
```

```python
# Task Θ(1) - Task 5: Insert and retrieve an item in a dictionary

def dict_insert_retrieve(d, key, value):
    d[key] = value
    return d[key]

# Prediction: Θ(1) average for insert and lookup

D = {}
print('\nDictionary insert & retrieve:')
print('Sample:', dict_insert_retrieve(D, 'k', 123))
print('Time (avg over 10000 runs):', time_callable(lambda: dict_insert_retrieve(D, 'k', 123), number=10000))
```

## Θ(log n) — Logarithmic Time Tasks

Implementations and tests below. Note: binary search requires a sorted list.

```python
# Task Θ(log n) - Task 1: Binary search on a sorted list of student IDs

def binary_search(L, target):
    lo, hi = 0, len(L)-1
    while lo <= hi:
        mid = (lo + hi) // 2
        if L[mid] == target:
            return True
        elif L[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return False

# Prediction: Θ(log n)

# Test sizes
sizes = [10**3, 10**4, 10**5, 10**6]
bs_times = []
for n in sizes:
    L = list(range(n))
    target = random.choice(L)
    t = time_callable(lambda: binary_search(L, target), number=100)
    bs_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, bs_times, label='Binary search', logscale=True, title='Binary search timing (log-log)')
```

```python
# Task Θ(log n) - Task 2: Guess a number between 1 and n using halving (simulate)

def guess_by_halving(n, secret):
    lo, hi = 1, n
    steps = 0
    while lo <= hi:
        steps += 1
        mid = (lo + hi) // 2
        if mid == secret:
            return True
        elif mid < secret:
            lo = mid + 1
        else:
            hi = mid - 1
    return False

# Prediction: Θ(log n)

sizes = [10**3, 10**4, 10**5, 10**6]
guess_times = []
for n in sizes:
    secret = random.randint(1, n)
    t = time_callable(lambda: guess_by_halving(n, secret), number=1000)
    guess_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, guess_times, label='Halving guess', logscale=True, title='Halving guess timing (log-log)')
```

```python
# Task Θ(log n) - Task 3: Find number of digits in a large integer

def num_digits(x):
    if x == 0:
        return 1
    return int(math.floor(math.log10(abs(x)))) + 1

# Iterative version (Θ(log n) where n = number magnitude)
def num_digits_iter(x):
    count = 0
    if x == 0:
        return 1
    x = abs(x)
    while x > 0:
        x //= 10
        count += 1
    return count

# Test with numbers of increasing digits
k_digits = [10, 100, 1000, 5000]
math_times = []
iter_times = []
for k in k_digits:
    num = 10**k - 1
    math_times.append(time_callable(lambda: num_digits(num), number=100))
    iter_times.append(time_callable(lambda: num_digits_iter(num), number=10))
    print(f'k={k} digits, math_time={math_times[-1]:.6e}, iter_time={iter_times[-1]:.6e}')

plot_results(k_digits, math_times, label='math.log10 method', logscale=True, title='Digits (math)')
plot_results(k_digits, iter_times, label='iterative method (dividing by 10)', logscale=True, title='Digits (iterative)')
```

```python
# Task Θ(log n) - Task 4: Simulate reducing a population by half until 1 remains

def reduce_half(n):
    steps = 0
    while n > 1:
        n = n // 2
        steps += 1
    return steps

# Prediction: Θ(log n)

sizes = [10**3, 10**4, 10**5, 10**6]
half_times = []
for n in sizes:
    t = time_callable(lambda: reduce_half(n), number=10000)
    half_times.append(t)
    print(f'n={n}, steps={reduce_half(n)}, time={t:.6e}s')

plot_results(sizes, half_times, label='Reduce by half', logscale=True, title='Reduce by half (timing)')
```

```python
# Task Θ(log n) - Task 5: Recursive division of n by 2 until 1

def rec_divide(n):
    if n <= 1:
        return 0
    return 1 + rec_divide(n//2)

# Prediction: Θ(log n)

sizes = [2**10, 2**15, 2**20]
rec_times = []
for n in sizes:
    t = time_callable(lambda: rec_divide(n), number=100)
    rec_times.append(t)
    print(f'n={n}, depth={rec_divide(n)}, time={t:.6e}s')

plot_results(sizes, rec_times, label='Recursive divide', logscale=True, title='Recursive divide timing')
```

## Θ(n) — Linear Time Tasks

Implementations and tests below.

```python
# Task Θ(n) - Task 1: Count words in a paragraph

def count_words(paragraph):
    return len(paragraph.split())

# Prediction: Θ(n)

paragraphs = ["word "+"word "*k for k in [100, 1000, 5000, 10000]]
word_times = []
for p in paragraphs:
    t = time_callable(lambda: count_words(p), number=1000)
    word_times.append(t)
    print(f'len(words)~{len(p.split())}, time={t:.6e}s')

plot_results([100,1000,5000,10000], word_times, label='Count words', title='Count words timing')
```

```python
# Task Θ(n) - Task 2: Find the max temperature in a list

def max_temperature(L):
    maximum = L[0]
    for x in L:
        if x > maximum:
            maximum = x
    return maximum

# Prediction: Θ(n)

sizes = [10**3, 10**4, 10**5]
max_times = []
for n in sizes:
    L = [random.random() for _ in range(n)]
    t = time_callable(lambda: max_temperature(L), number=10)
    max_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, max_times, label='Max temperature', logscale=True, title='Max temperature timing')
```

```python
# Task Θ(n) - Task 3: Check if a customer ID is in a list

def contains_id(L, key):
    for item in L:
        if item == key:
            return True
    return False

# Prediction: Θ(n)

sizes = [10**3, 10**4, 10**5]
contains_times = []
for n in sizes:
    L = list(range(n))
    key = random.choice(L)
    t = time_callable(lambda: contains_id(L, key), number=100)
    contains_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, contains_times, label='Contains ID', logscale=True, title='Contains ID timing')
```

```python
# Task Θ(n) - Task 4: Compute sum of grades for all students

def sum_grades(L):
    total = 0
    for x in L:
        total += x
    return total

# Prediction: Θ(n)

sizes = [10**3, 10**4, 10**5]
sum_times = []
for n in sizes:
    L = [random.random() for _ in range(n)]
    t = time_callable(lambda: sum_grades(L), number=10)
    sum_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, sum_times, label='Sum grades', logscale=True, title='Sum grades timing')
```

```python
# Task Θ(n) - Task 5: Print all items of a shopping list

def print_shopping(L):
    for item in L:
        _ = item  # simulate printing but avoid console I/O

# Prediction: Θ(n)

sizes = [10, 100, 1000, 5000]
print_times = []
for n in sizes:
    L = [f'item{i}' for i in range(n)]
    t = time_callable(lambda: print_shopping(L), number=100)
    print_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, print_times, label='Print shopping list', title='Print shopping timing')
```

## Θ(n log n) — Log-Linear Time Tasks

Implementations and tests below.

```python
# Task Θ(n log n) - Task 1: Implement merge sort

def merge(left, right):
    i = j = 0
    out = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            out.append(left[i]); i += 1
        else:
            out.append(right[j]); j += 1
    out.extend(left[i:])
    out.extend(right[j:])
    return out

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

# Prediction: Θ(n log n)

sizes = [1000, 2000, 4000]
ms_times = []
for n in sizes:
    arr = [random.random() for _ in range(n)]
    t = time_callable(lambda: merge_sort(arr), number=1)
    ms_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, ms_times, label='Merge sort', logscale=True, title='Merge sort timing')
```

## Θ(n²) — Polynomial (Quadratic) Tasks

Implementations and tests below. Note that quadratic tasks grow quickly; keep `n` moderate.

```python
# Task Θ(n²) - Task 1: Compare every student with every other for duplicate IDs

def has_duplicate_pairs(L):
    n = len(L)
    for i in range(n):
        for j in range(i+1, n):
            if L[i] == L[j]:
                return True
    return False

# Prediction: Θ(n^2)

sizes = [200, 400, 800]
dup_times = []
for n in sizes:
    L = list(range(n))
    L[-1] = 0
    t = time_callable(lambda: has_duplicate_pairs(L), number=5)
    dup_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, dup_times, label='Duplicate pairs', logscale=True, title='Duplicate pairs timing')
```

```python
# Task Θ(n²) - Task 2: Generate an n x n multiplication table

def multiplication_table(n):
    table = [[i*j for j in range(n)] for i in range(n)]
    return table

# Prediction: Θ(n^2)

sizes = [100, 200, 400]
mt_times = []
for n in sizes:
    t = time_callable(lambda: multiplication_table(n), number=3)
    mt_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, mt_times, label='Multiplication table', logscale=True, title='Multiplication table timing')
```

```python
# Task Θ(n²) - Task 3: Compute pairwise distances between cities

def pairwise_distances(points):
    n = len(points)
    res = []
    for i in range(n):
        for j in range(i+1, n):
            dx = points[i][0]-points[j][0]
            dy = points[i][1]-points[j][1]
            res.append(math.hypot(dx, dy))
    return res

# Prediction: Θ(n^2)

sizes = [100, 200, 400]
pw_times = []
for n in sizes:
    pts = [(random.random(), random.random()) for _ in range(n)]
    t = time_callable(lambda: pairwise_distances(pts), number=3)
    pw_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, pw_times, label='Pairwise distances', logscale=True, title='Pairwise distances timing')
```

```python
# Task Θ(n²) - Task 4: Check for duplicate pairs in a list (simple pairwise check)

def check_duplicate_pairs(L):
    n = len(L)
    for i in range(n):
        for j in range(i+1, n):
            if L[i] == L[j]:
                return True
    return False

# Prediction: Θ(n^2)

sizes = [200, 400, 800]
cdp_times = []
for n in sizes:
    L = list(range(n))
    L[-1] = 0
    t = time_callable(lambda: check_duplicate_pairs(L), number=5)
    cdp_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, cdp_times, label='Check duplicate pairs', logscale=True, title='Check duplicate pairs timing')
```

```python
# Task Θ(n²) - Task 5: Brute-force check all classroom seat pairs

def seat_pairs(seats):
    pairs = []
    n = len(seats)
    for i in range(n):
        for j in range(i+1, n):
            pairs.append((seats[i], seats[j]))
    return pairs

# Prediction: Θ(n^2)

sizes = [100, 200, 400]
seat_times = []
for n in sizes:
    seats = list(range(n))
    t = time_callable(lambda: seat_pairs(seats), number=3)
    seat_times.append(t)
    print(f'n={n}, time={t:.6e}s')

plot_results(sizes, seat_times, label='Seat pairs', logscale=True, title='Seat pairs timing')
```

## Θ(2^n) — Exponential Tasks

These tasks blow up quickly; use small `n`.

```python
# Task Θ(2^n) - Task 1: Generate all subsets of a set of size n

def all_subsets(s):
    res = [[]]
    for elem in s:
        res += [curr + [elem] for curr in res]
    return res

# Prediction: Θ(2^n)

sizes = [10, 12, 14]
subset_times = []
for n in sizes:
    s = list(range(n))
    t = time_callable(lambda: all_subsets(s), number=3)
    subset_times.append(t)
    print(f'n={n}, subsets={2**n}, time={t:.6e}s')

plot_results(sizes, subset_times, label='All subsets', title='All subsets timing (exponential)')
```

```python
# Task Θ(2^n) - Task 2: Towers of Hanoi recursively

def hanoi_moves(n, src='A', dst='C', aux='B', record=False):
    if n == 0:
        return [] if record else 0
    if n == 1:
        if record:
            return [(src, dst)]
        return 1
    if record:
        return hanoi_moves(n-1, src, aux, dst, record) + [(src,dst)] + hanoi_moves(n-1, aux, dst, src, record)
    return 2*hanoi_moves(n-1, src, dst, aux, record) + 1

# Prediction: Θ(2^n)

sizes = [10, 12, 14]
hanoi_times = []
for n in sizes:
    t = time_callable(lambda: hanoi_moves(n, record=False), number=3)
    hanoi_times.append(t)
    print(f'n={n}, moves={2**n - 1}, time={t:.6e}s')

plot_results(sizes, hanoi_times, label='Hanoi moves', title='Hanoi timing (exponential)')
```

```python
# Task Θ(2^n) - Task 3: Naive recursive Fibonacci

def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# Prediction: Θ(2^n)

sizes = [20, 25, 30]
fib_times = []
for n in sizes:
    t = time_callable(lambda: fib(n), number=1)
    fib_times.append(t)
    print(f'n={n}, fib(n) computed, time={t:.6e}s')

plot_results(sizes, fib_times, label='Naive fib', title='Naive Fibonacci timing (exponential)')
```

```python
# Task Θ(2^n) - Task 4: Generate all binary strings of length n

def all_bin_strings(n):
    if n == 0:
        return ['']
    smaller = all_bin_strings(n-1)
    return [s + '0' for s in smaller] + [s + '1' for s in smaller]

# Prediction: Θ(2^n)

sizes = [12, 14, 16]
bin_times = []
for n in sizes:
    t = time_callable(lambda: all_bin_strings(n), number=3)
    bin_times.append(t)
    print(f'n={n}, count={2**n}, time={t:.6e}s')

plot_results(sizes, bin_times, label='All binary strings', title='All binary strings timing (exponential)')
```

```python
# Task Θ(2^n) - Task 5: Simulate trying all password combinations of length n (digits only)

def all_passwords(n):
    if n == 0:
        return ['']
    smaller = all_passwords(n-1)
    out = []
    for s in smaller:
        for d in '0123456789':
            out.append(s + d)
    return out

# Prediction: Θ(10^n) which is exponential

sizes = [4, 5, 6]
pass_times = []
for n in sizes:
    t = time_callable(lambda: all_passwords(n), number=1)
    pass_times.append(t)
    print(f'n={n}, combinations={10**n}, time={t:.6e}s')

plot_results(sizes, pass_times, label='All passwords (digits)', title='All passwords timing (exponential)')
```
